```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python
def automated_crypto_pull():
    url = 'https://coinmarketcap.com/currencies/bitcoin/'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    crypto_name = soup.find('span', class_='sc-c1554bc0-0 hhUhBo').text
    crypto_price = soup.find('span', class_="sc-c1554bc0-0 RbQXx base-text").text
    final_price = crypto_price.replace('$','').replace(',','')
    
    date_time = datetime.now()
    
    dict = {
        'Crypto Name': crypto_name,
        'Price': final_price,
        'TimeStamp': date_time
    }
    
    df = pd.DataFrame([dict])
    
    file_path = r'C:\Users\Areej\OneDrive\Documents\Python\Example folder\Crypto_Automated_pull.csv'
    
    if os.path.exists(file_path):
        df.to_csv(file_path, mode='a', header=False, index=False)
    else:
        df.to_csv(file_path, index=False)
    
    print(df)
```

```python
while True:
    automated_crypto_pull()
    time.sleep(10)
```
